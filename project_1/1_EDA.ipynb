```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
df=pd.read_csv('https://lukeb.co//python_csv')
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
import ast 
df['job_skills'] = df['job_skills'].apply(
    lambda skill: ast.literal_eval(skill) if isinstance(skill, str) else skill
)
```

```python
import seaborn as sns
```

```python
# filter for usa 
df_US=df[(df['job_country']== 'United States') & (df['job_title_short']== 'Data Analyst')].copy()
```

## pick columns in order to pefomance some intial EDA

```python

df_plot=df_US['job_location'].value_counts().head(10).to_frame()
#Anywhere correlates to other remote jobs,turn into datafram

sns.set_theme(style='ticks') 

sns.barplot(data=df_plot,x='count' ,y='job_location',hue='count',palette='dark:b_r',legend=False)
sns.despine()#to remove border 

plt.title('Number of jobs per job title')
plt.xlabel('numbers of Jobs')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
fig, ax= plt.subplots(1,3,figsize=(10,10))

#values of dictionary represent what i want as header on the pie
dict_column={'job_work_from_home':'work from home', 
    'job_no_degree_mention':'job deg Req',
    'job_health_insurance':'help insurance offered'}

#column gets the DataFrame column name, title gets the display title
for i, (column,title) in enumerate(dict_column.items()):
    ax[i].pie(df_US[column].value_counts(), startangle=90, autopct='%1.1f%%',labels=['False','True'])
    ax[i].set_title(title)
#labels=['False','True'] sets text labels for slices (applies in the order of the sizes passed)

plt.show()
```

```python

df_plot=df_US['company_name'].value_counts().head(10).to_frame()
#Anywhere correlates to other remote jobs,turn into datafram
plt.figure(figsize=(10, 6))
sns.set_theme(style='ticks') 

sns.barplot(data=df_plot,x='count' ,y='company_name',hue='count',palette='dark:b_r',legend=False)
sns.despine()#to remove border 

plt.title('count of companies for data analysis in united states')
plt.xlabel('numbers of Jobs')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
