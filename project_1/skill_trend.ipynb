```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
df=pd.read_csv('https://lukeb.co//python_csv')
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
import ast 
df['job_skills'] = df['job_skills'].apply(
    lambda skill: ast.literal_eval(skill) if isinstance(skill, str) else skill
)
```

```python
import seaborn as sns
```

```python

# "Skill trending" means:
# Which skills are becoming more (or less) popular over time in job postings.
```

```python
df_DA_US=df[(df['job_title_short']== 'Data Analyst') & (df['job_country']=='United States')].copy()
```

```python
df_DA_US['job_posted_month_no']=df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US.head(2)
```

```python
df_DA_US_explode=df_DA_US.explode('job_skills')
```

```python
#count driffent skills 
df_DA_US_pivot=df_DA_US_explode.pivot_table(index='job_posted_month_no',columns='job_skills',aggfunc='size',fill_value=0)
df_DA_US_pivot
#size applies to the combination of index and columns.
```

```python
df_DA_US_pivot.loc['total']=df_DA_US_pivot.sum()
```

```python
df_DA_US_pivot=df_DA_US_pivot[df_DA_US_pivot.loc['total'].sort_values(ascending=False).index]
```

```python
df_DA_US_pivot=df_DA_US_pivot.drop('total')
```

```python
df_DA_US_pivot
```

## we need to use percentage 

```python
Data_analysis_total=df_DA_US.groupby('job_posted_month_no').size()
#.size() counts total job postings per month
Data_analysis_total
```

```python
df_DA_US_percent=df_DA_US_pivot.div(Data_analysis_total/100,axis=0)
df_DA_US_percent
```

```python
df_DA_US_percent=df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = pd.to_datetime(
    df_DA_US_percent['job_posted_month_no'], format='%m').dt.strftime('%b')
```

```python
df_DA_US_percent=df_DA_US_percent.set_index('job_posted_month')

```

```python
df_DA_US_percent=df_DA_US_percent.drop(columns='job_posted_month_no')
```

```python
df_DA_US_percent
```

```python
df_plot=df_DA_US_percent.iloc[:,:5]
df_plot
```

```python

# plt.Figure(figsize=(12,22))
sns.lineplot(data=df_plot,dashes=False,palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('trending top skills for data analysts in the US')
plt.xlabel('2023')
plt.ylabel('likehood in job postng')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax=plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))


for i in range(5):
    plt.text(12, df_plot.iloc[-1, i] + i * -0.95, df_plot.columns[i])

plt.savefig('skills_trend.png')
plt.show()
```
