```python
# we want to convert it to list :Use ast.literal_eval():
import ast #convert datatype to what actual need to be
#type(ast.literal_eval(df['job_skills'][1]))

```

```python
#clean this column used apply function
def clean_list(skill):
    if pd.notna(skill):
        return ast.literal_eval(skill)
    
    
df['job_skills'] = df['job_skills'].apply(clean_list)
```

## summary

```python
#syntax:lambda arguments: expression
#pd.notna()` is the **opposite** of `pd.isna()`.pd.isna()` → **True** for missing values
##pd.notna()` → **True** for non-missing values
#syntax:lambda x: value_if_true if condition else value_if_false
#instance() built-in Python function that checks if skill is a string to ensure ast.literal_eval() is only applied to string values
#syntax:instance(object, type)
#f isinstance(skill, str) is True (i.e., skill is a string), ast.literal_eval(skill) is executed to convert the string into a object.
#if :isinstance(skill, str): Convert the string skill to a Python object (e.g., a list) 
#The else part returns the original skill value without any modification if skill is not a string.

#summary:
#If skill is a string: Convert it using ast.literal_eval(skill).
#If skill is not a string: Return skill as-is, not "nothing."
```
