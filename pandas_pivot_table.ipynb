## we used pivot tables to analyze trends over time 
## syntax :df.pivot table(value,index,columns,aggfunc)

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
df=pd.read_csv('https://lukeb.co//python_csv')
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
#i want to get count of job titles 
df.pivot_table(values='salary_year_avg', index='job_title_short',aggfunc='median',columns='job_country')
#index mean group by rows and coulmn mean group by columns ,values=column to aggregate 
```

```python
#i want smaller table 
df.pivot_table(values='salary_year_avg', index='job_country',aggfunc='median',columns='job_title_short')
#each value in table represent median of salary for specific job in specific country
```

```python
df_job_country_salary=df.pivot_table(values='salary_year_avg', index='job_country',
aggfunc='median',columns='job_title_short')
```

```python
# i want access to 6 countries
```

```python
top_countries=df['job_country'].value_counts().sort_values(ascending=False).head(6).index
top_countries #this code gives the top 6 countries that appear most o in `job_country` column 
```

```python

 #we use loc to access to these top countries
#loc[] must have row part but column part is optional.
#we get index of top_countries that mean we get rows and use rows in loc method
df_job_country_salary=df_job_country_salary.loc[top_countries] #to save changes
df_job_country_salary
```

```python

job_titles=['Data Analyst','Data Engineer','Data Scientist']
#you are using the list job_titles to select columns from your DataFrame.
df_job_country_salary=df_job_country_salary[job_titles]
df_job_country_salary
```

```python
df_job_country_salary.plot(kind='bar')
plt.xlabel('')
plt.ylabel('median ($)')
plt.title('median salary by country and job title')
plt.xticks(rotation=45,ha='right')
plt.show()
```
