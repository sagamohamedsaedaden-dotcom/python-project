```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
df=pd.read_csv('https://lukeb.co//python_csv')
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
import ast
df['job_skills'] = df['job_skills'].apply(
    lambda skill: ast.literal_eval(skill) if isinstance(skill, str) else skill
)
```

```python
df.head(5)
```

```python
# i want precentage 
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90, autopct='%1.1f%%')
plt.title('work from home status')
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('jop titles in dataset')
plt.ylabel('')
plt.show()
```

```python
fig, ax= plt.subplots(1,3,figsize=(10,10))

#values of dictionary represent what i want as header on the pie
dict_column={'job_work_from_home':'work from home', 
    'job_no_degree_mention':'job deg Req',
    'job_health_insurance':'help insurance offered'}


#column gets the DataFrame column name, title gets the display title
for i, (column,title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%',labels=['False','True'])
    ax[i].set_title(title)
#labels=['False','True'] sets text labels for slices (applies in the order of the sizes passed)

plt.show()

```

```python
#https://docs.python.org/3/library/string.html
#string format (f):float number,For a given precision p,formats the number as a decimal number with 
#exactly  digits following the decimal point.
#autopct='%f%%'shows full float numbers(many decimals)like: 23.456789%, %1.1f%%'→ rounded to 1 decimal


#→ df[column] tells pandas to look inside df for a column 
#if you wrote only column.value_counts(), it would cause an error.
# in for loop comprehision expression and iterable can not be optional 
```
