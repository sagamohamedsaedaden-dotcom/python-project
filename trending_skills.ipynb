## trends of skills throuth year for Data Analyst

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
df=pd.read_csv('https://lukeb.co//python_csv')
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
import ast
df['job_skills'] = df['job_skills'].apply(
    lambda skill: ast.literal_eval(skill) if isinstance(skill, str) else skill
)
```

## i want to focus on Data Analyst

```python
df_da=df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_da
```

```python
# want to aggregate skills on monthly basis 
df_da['job_posted_month']=df_da['job_posted_date'].dt.month
```

```python
df_da_explode=df_da.explode('job_skills')
df_da_explode
```

```python
#`size` tells us the count of rows for each combination of `job_posted_month` and `job_skills`.
df_da_pivot=df_da_explode.pivot_table(index='job_posted_month', columns='job_skills', aggfunc='size',fill_value=0)
# fill_value=0 mean put 0 instead of non becaue we canot plot non values
```

```python
df_da_pivot
```

```python
df_da_pivot.loc['total']= df_da_pivot.sum()
#That adds a row named 'total'with the sum of each column
```

```python
df_da_pivot
```

```python
df
```

```python
# we want to sort columns by total row
df_da_pivot.loc['total'].sort_values(ascending=False)
```

```python
#we want index not value 
df_da_pivot.loc['total'].sort_values(ascending=False).index
```

```python
df_da_pivot[df_da_pivot.loc['total'].sort_values(ascending=False).index]
# say to pandas to select columns from df_da_pivot based on ordered list of column names
#This returns the same table but with columns reordered 
```

```python
df_da_pivot=df_da_pivot[df_da_pivot.loc['total'].sort_values(ascending=False).index]
df_da_pivot=df_da_pivot.drop('total')
```

```python
df_da_pivot
```

```python
df_da_pivot.iloc[:,:5]
```

```python
df_da_pivot.iloc[:,:5].plot(kind='line')
plt.title('top skills for data analyst per month') #now i know why use head(top) instead of head(10)
plt.xlabel('')
plt.ylabel('count')
plt.show()
```

## another way to plot 


```python
df_da_pivot=df_da_pivot.reset_index()
df_da_pivot['job_posted_month']=df_da_pivot['job_posted_month'].apply(lambda x:pd.to_datetime (x,format='%m').strftime('%B'))
```

```python
df_da_pivot=df_da_pivot.set_index('job_posted_month')
```

```python
df_da_pivot.iloc[:,:5].plot(kind='line')
plt.title('top skills for data analyst per month') #now i know why use head(top) instead of head(10)
plt.xlabel('')
plt.ylabel('count')
plt.show()
```
